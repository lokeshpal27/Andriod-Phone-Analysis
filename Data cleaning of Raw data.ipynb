```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\hp\\OneDrive\\Desktop\\projects\\moblile phone\\mobile phone price prediction.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    if df[i].dtype != 'O':
        df[i].fillna(df[i].mean(numeric_only=True), inplace=True)
    else:
        if not df[i].mode().empty:
            df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['fast_charging'] = df['fast_charging'].str.extract('(\d+)').astype(float)
```

```python
df['Inbuilt_memory'] = df['Inbuilt_memory'].str.extract('(\d+)').astype(float)
```

```python
df['Price'] = df['Price'].str.replace(',',"").astype(int)
```

```python
df.head(3)
```

```python
df['Model'] = df['Name'].str.split(' ', 1).str[1]
```

```python
df.drop(columns='Name',axis= 1)
```

```python
df['Battery'] = df['Battery'].str.extract('(\d+)').astype(int)
```

```python
df['Ram'] = df['Ram'].str.extract('(\d+)').astype(float)
```

```python
df['Ram']=df['Ram'].astype(int)
```

```python
df['Display'] = df['Display'].str.extract('(\d+)').astype(float)
```

```python
df.drop(df.columns[0],axis=1, inplace= True)
```

```python
df['External_Memory']=df['External_Memory'].str.extract('(\d+)').astype(float)
```

```python
df['External_Memory']=df['External_Memory'].replace(1.0, 1000.0)
```

```python
df['Screen_pixels'] = df['Screen_resolution'].str.extract(r'(\d+ x \d+ px)')
```

```python
df.drop(columns='Screen_resolution',axis=1 ,inplace= True)
```

```python
df.drop(columns='Name',axis=1 ,inplace= True)
```

```python
df['5G_Support'] = np.where(df['No_of_sim'].str.contains('5G'), 'Yes', 'No')
```

```python
df.head(4)
```

```python
df[['Rear_Camera', 'Front_Camera']] = df['Camera'].str.split('&', expand=True)
```

```python
df['Rear_Camera'] = df['Rear_Camera'].str.extract('(\d+)').astype(float)
df['Front_Camera'] = df['Front_Camera'].str.extract('(\d+)').astype(float)
```

```python
df.drop(columns=['No_of_sim','Camera'],axis=1,inplace=True)
```
